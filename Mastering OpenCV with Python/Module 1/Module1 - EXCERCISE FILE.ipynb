```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from IPython.display import Image
plt.rcParams['image.cmap'] = 'gray'
```

### READING, DISPLAYING and SAVING Image

```python
img_1 = cv2.imread('Apollo-11-launch.jpg', 0)
print(img_1.shape)
plt.figure(figsize=[10,10])
plt.imshow(img_1,cmap = 'gray')
cv2.imwrite('PNG.png',img_1)
```

### COLOR IMAGES

```python
Image_1 = cv2.imread('Emerald_Lakes_New_Zealand.jpg', cv2.IMREAD_COLOR)
print('imgae size is', Image_1.shape)
log1 = Image_1[:,:, ::-1]

cv2.imwrite('image.jpg', log1)

chg = cv2.cvtColor(log1, cv2.COLOR_BGRA2GRAY)
print('shape is', chg.shape)
plt.figure(figsize=[10,10])
plt.imshow(chg)
```

### BASIC IMAGE MANIPULATION

```python
imge = cv2.imread('New_Zealand_Boat.jpg', cv2.IMREAD_COLOR)
plt.figure(figsize=[3,3])
plt.imshow(imge[:,:,::-1])

croping = imge[200:300, 300:420]
plt.imshow(croping[:,:,::-1])

Resize1 = cv2.resize(croping,None,fx=2,fy=2)
plt.imshow(croping[:,:,::-1])

image_n = cv2.flip(Resize1,1)
plt.figure(figsize=[20,5])
plt.subplot(141), plt.imshow(image_n[:,:,::-1]); plt.title('horizontal')
```

### ANNOTATING IMAGES

```python
#define colors for annotations
yellow = (0,255,255)
red = (0,0,255)
magneta = (255,0,255)
green = (0,255,0)
```

```python
image1 = cv2.imread('Apollo-11-launch.jpg', cv2.IMREAD_COLOR)
text = 'Apollo 1 saturn V launch july 16,1969'
font_face = cv2.FONT_HERSHEY_PLAIN
font_scale = 1.2
font_color = green
font_thick = 2

image1 = cv2.putText(image1, text, (10, 50), font_face, font_scale, font_color, font_thick, cv2.LINE_AA)
# plt.figure(figsize=[12,12])
# plt.imshow(image2[:,:,::-1])

image1 = cv2.rectangle(image1, (500,100),(700,600),magneta, thickness=3, lineType=cv2.LINE_AA)

plt.figure(figsize=[12,12])
plt.imshow(image1[:,:,::-1])
```
